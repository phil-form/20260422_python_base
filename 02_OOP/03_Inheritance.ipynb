# Inheritance

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Cat(Animal):
    def __init__(self, name, age, race):
        super().__init__(name, age)
        self.race = race

class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)
        self.breed = breed

d = Dog("Lala", 10, "German-shepherd")
c = Cat("Jean-Paul", 3, "Siamese")
print(d.name)
print(d.age)
print(d.breed)

print(c.name)
print(c.age)
print(c.race)
```

```python
def print_animal_info(animal: Animal):
    print(animal.__class__.__name__, " " , animal.name, " ", animal.age)

print_animal_info(d)
print_animal_info(c)
```

```python
class Refuge:
    def __init__(self):
        self.animals = []

    def add_animal(self, animal: Animal):
        self.animals.append(animal)

    def print_animals(self):
        print("Animal list : ")
        for animal in self.animals:
            print_animal_info(animal)

refuge = Refuge()
refuge.print_animals()
refuge.add_animal(d)
refuge.add_animal(c)
refuge.print_animals()
```

# Overriding



```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def sleep(self):
        print("Sleeping")

class Cat(Animal):
    def __init__(self, name, age, race):
        super().__init__(name, age)
        self.race = race

    def sleep(self):
        print("Sleeping but twice as much")

class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)
        self.breed = breed
```

```python
d = Dog("Lala", 10, "German-shepherd")
c = Cat("Jean-Paul", 3, "Siamese")

d.sleep()
```

```python
c.sleep()
```

```python
def make_animal_sleep(animal: Animal):
    animal.sleep()

make_animal_sleep(c)
```

# Abstract class

```python
from abc import ABC, abstractmethod

class BankAccount(ABC):
    def __init__(self, amount):
        self.amount = amount

    def deposit(self, amount):
        self.amount += amount

    @abstractmethod
    def withdraw(self, amount):
        pass

class SavingsAccount(BankAccount):
    pass

s = SavingsAccount(500)
```

```python
class SavingsAccount(BankAccount):
    def withdraw(self, amount):
        self.amount -= amount

s = SavingsAccount(500)
print(s.amount)
s.withdraw(500)
print(s.amount)
```
