# Generators

```python
my_list = [i for i in range(1_000_000)]

for el in my_list:
    pass

def gen():
    for i in range(1_000_000):
        yield i

for el in gen():
    pass
```

```python
def file_lines(file_path):
    with open(file_path) as f:
        for line in f:
            yield line

for line in file_lines("data/test_pandas.csv"):
    print(line)
```
