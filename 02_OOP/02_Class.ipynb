# Attributes

```python
class Person:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name

p1 = Person("John", "Doe")
p2 = Person("Jane", "Doe")
print(p1.first_name)
print(p2.first_name)

print("John lastname")
print(p1.last_name)
p1.last_name = "Asdf"
print(p1.last_name)
```

# Add attributes

The wrong way to add an attribute would be to do as such :

```python
p1.age = 41
print(p1.age)
```

As you can see the problem is that the age attribute is not common to all objects of class Person.

```python
print(p1.__dict__)
print(p2.__dict__)
```

The right way to do is to use the constructor

```python
class Person2:
    def __init__(self, first_name, last_name, age):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age

# I can also set default values in the constructor :
class Person3:
    def __init__(self, first_name, last_name, age=1):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age

p3 = Person3("John", "Doe")
print(p3.first_name)
print(p3.last_name)
print(p3.age)
```

# Methods

```python
class Car:
    def __init__(self, make, model, year, speed = 0):
        self.make = make
        self.model = model
        self.year = year
        self.speed = speed

    def accelerate(self):
        self.speed += 50

    def brake(self):
        self.speed -= 50

car1 = Car("asdf", "asdf", 2026)

print(car1.speed)
car1.accelerate()
print(car1.speed)
car1.brake()
print(car1.speed)
```
