# Add attributes

The wrong way to add an attribute would be to do as such :

````python
p1.age = 41
print(p1.age)
````

As you can see the problem is that the age attribute is not common to all objects of class Person.

````python
print(p1.__dict__)
print(p2.__dict__)
````

The right way to do is to use the constructor

````python
class Person2:
    def __init__(self, first_name, last_name, age):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age

# I can also set default values in the constructor :
class Person3:
    def __init__(self, first_name, last_name, age=1):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age

p3 = Person3("John", "Doe")
print(p3.first_name)
print(p3.last_name)
print(p3.age)
````

# Encapsulation

## General principle

````python
class Person():
    def __init__(self, first_name, last_name):
        # I can __ to make the parameter private
        # That way the parameter cannot be accessed (in read or write)
        # From outside the class
        self.__first_name = first_name
        self.__last_name = last_name

    # I can create a getter using @property
    # This will allow to access the value of __first_name through the parameter first_name
    @property
    def first_name(self):
        return self.__first_name

    # I can create a setter using @parameter_name.setter
    # This will allow me to change the value of the encapsuled parameter
    @first_name.setter
    def first_name(self, new_first_name):
        self.__first_name = new_first_name

    # I can set a parameter as read only (so no setters)
    @property
    def last_name(self):
        return self.__last_name

    # I can also create a parameter, that will reckon the value during execution
    @property
    def full_name(self):
        return f"{self.first_name} {self.last_name}"

p = Person("John", "Doe")
print(p.first_name)
print(p.last_name)
print(p.full_name)

p.last_name = "Asdf"
````

## Generated properties

````python
class Rectangle:
    """
    My custom doc string
    """
    def __init__(self, width, height):
        self.width = width
        self.height = height

    @property
    def area(self):
        return self.width * self.height

    @property
    def perimeter(self):
        return (self.width + self.height) * 2

rec = Rectangle(2, 3)
print(rec.area)
print(rec.perimeter)
````

# Data Model

```
__str__ and __repr__

````python
class Student:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name

    def __str__(self):
        return f"{self.first_name} {self.last_name}"

    def __repr__(self):
        return f"REPR {self.first_name} {self.last_name}"

st = Student("John", "Doe")
print(st)
print(str(st))
print(f"{st}")
print(f"{st!r}")
````

__len__ and __add__
```

````python
class Library():
    def __init__(self, books):
        self.books = books

    def __len__(self):
        return len(self.books)

lib = Library(["book1", "book2", "book3"])
print(len(lib))
````

````python
class Library():
    def __init__(self, books):
        self.books = books

    def __len__(self):
        return len(self.books)

    def __add__(self, other: Library):
        return Library(self.books + other.books)

lib = Library(["book1", "book2", "book3"])
print(len(lib))

lib2 = Library(["book1", "book2", "book3"])

lib3 = lib + lib2
print(lib3.books)
````

```
__eq__  => ==
__ne__  => !=
__lt__  => <
__le__  => <=
__gt__  => >
__ge__  => >=
```

````python
class Student:
    def __init__(self, first_name, last_name, avg_grade):
        self.first_name = first_name
        self.last_name = last_name
        self.avg_grade = avg_grade

    def __eq__(self, other):
        return self.avg_grade == other.avg_grade

    def __lt__(self, other):
        return self.avg_grade < other.avg_grade

    def __le__(self, other):
        return self.avg_grade <= other.avg_grade

    def __gt__(self, other):
        return self.avg_grade > other.avg_grade

    def __ge__(self, other):
        return self.avg_grade >= other.avg_grade

s1 = Student("John", "Doe", 10)
s2 = Student("James", "asdf", 20)
print(s1.__dict__)
print(s2.__dict__)
print(s1 == s2)
print(s1 != s2)
print(s1 < s2)
print(s1 <= s2)
print(s1 > s2)
print(s1 >= s2)
````

```
__getitem__ => val = obj[n]
__setitem__ => obj[n] = val
__iter__ => for val in obj:
```

````python
class Library():
    def __init__(self, books):
        self.books = books
        self.iter_pos = 0

    def __len__(self):
        return len(self.books)

    def __getitem__(self, item):
        return self.books[item]

    def __setitem__(self, key, value):
        self.books[key] = value

    # The __iter__ function is called during the initialization of the for loop.
    def __iter__(self):
        # If I want to use the existing iterator for the for loop
        # return iter(self.books)
        self.iter_pos = 0
        return self

    # To create a custom iterator
    # The __next__ function is called at each pass of the for loop.
    def __next__(self):
        if self.iter_pos < len(self.books):
            item = self.books[self.iter_pos]
            self.iter_pos += 1
            return item

        raise StopIteration


    def __add__(self, other: Library):
        return Library(self.books + other.books)
````
