# Basic File Reading

```python
# Open a file in python
file = open("data/test_pandas.csv", "r")

print(file.read())
# Pay attention to close the file once you're done.
file.close()
```

## Safer open

Using with open(...) as file:

Python will automaticaly close the file at the end of the block

```python
with open("data/test_pandas.csv", "r") as f:
    print(f.read())
```

## Read line by line

```python
with open("data/test_pandas.csv", "r") as f:
    for line in f:
        print(line)
```

```python
with open("data/test_pandas.csv", "r") as f:
    for line in f:
        # Strip the \n at the end of the line
        print(line.strip())
```

# Parse data

```python
data = []

with open("data/test_pandas.csv", "r") as f:
    for line in f:
        # Strip the \n at the end of the line
        line = line.strip()
        name, age, city = line.split(",")
        data.append({
            "name": name,
            "age": int(age),
            "city": city
        })

print(data)
```

```python
data = []

with open("data/test_pandas.csv", "r") as f:
    # Skip the header line
    next(f)
    for line in f:
        # Strip the \n at the end of the line
        line = line.strip()
        name, age, city = line.split(",")
        data.append({
            "name": name,
            "age": int(age),
            "city": city
        })

print(data)
```

# Writing

```python
data = [{'name': 'Alice', 'age': 30, 'city': 'New York'}, {'name': 'Bob', 'age': 25, 'city': 'Los Angeles'}, {'name': 'Charlie', 'age': 35, 'city': 'Chicago'}]

with open("data/output.csv", "w") as f:
    for person in data:
        line = f"{person['name']},{person['age']},{person['city']}\n"
        f.write(line)
```

# Append

Write -> rewrite the whole file, so it does not keep the original content

Append -> add at the end of the file

```python
with open("data/output.csv", "a") as f:
    f.write("bob,30,Brussels")
```

# Read the whole file

```python
with open("data/output.csv", "r") as f:
    print(f.read())
```

```python
with open("data/output.csv", "r") as f:
    print(f.readlines())
```

# Example of data processing

```python
with open("data/output.csv", "r") as f:
    total_age = 0
    count = 0

    for line in f:
        age = line.strip().split(",")[1]
        total_age += int(age)
        count += 1

    average = total_age / count
    print(average)
```

# Handle errors

```python
try:
    with open("data/notexist.csv", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("Not found")
```

# Json

```python
import json

data = []
with open("data/output.csv", "r") as f:
    for line in f:
        name, age, city = line.strip().split(',')

        data.append({
            "name": name,
            "age": int(age),
            "city": city
        })

# Write JSON
with open("data/output.json", "w") as f:
    json.dump(data, f)

# Read JSON
with open("data/output.json", "r") as f:
    print(json.load(f))
```
