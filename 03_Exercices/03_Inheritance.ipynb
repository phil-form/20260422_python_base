1. Basic inheritance

Create:

* a class `Animal` with attribute `name`
* a method `speak()` that prints `"Some sound"`

Create a subclass `Dog` that inherits from `Animal` and uses the same method.

2. Method overriding

Modify `Dog` so that:

* `speak()` prints `"Woof!"` instead of `"Some sound"`

3. Add new attribute in child

Create:

* `Animal(name)`
* `Cat(name, color)`

The `Cat` class should:

* call the parent constructor
* store `color`

4. Use `super()`

Create:

* class `Person(name)`
* class `Student(name, grade)`

Use `super()` to initialize `name`.

5. Extend a method

Create:

* class `Vehicle` with method `info()` Ôćĺ prints `"This is a vehicle"`

Create `Car` that:

* calls parent `info()`
* then prints `"This is a car"`

6. Multiple methods inheritance

Create:

* `Shape` with method `describe()` Ôćĺ `"I am a shape"`

Create `Rectangle`:

* attributes: width, height
* method `area()`

7. Override + new behavior

Create:

* `Employee(name, salary)`
* method `work()` Ôćĺ `"Working..."`

Create `Manager`:

* override `work()` Ôćĺ `"Managing team"`

8. Add functionality to parent method

Create:

* `BankAccount(balance)`
* method `deposit(amount)`

Create `SavingsAccount`:

* inherits from `BankAccount`
* adds method `add_interest(rate)`
* increases balance using interest

9. Polymorphism with inheritance

Create:

* `Animal` with method `speak()`

Create subclasses:

* `Dog` Ôćĺ `"Woof"`
* `Cat` Ôćĺ `"Meow"`
* `Cow` Ôćĺ `"Moo"`

Create a list of animals and loop through them calling `speak()`.
