1. Basic inheritance

Create:

* a class `Animal` with attribute `name`
* a method `speak()` that prints `"Some sound"`

Create a subclass `Dog` that inherits from `Animal` and uses the same method.

```python
class Animal:
    def __init__(self, name: str):
        self.name = name

    def speak(self):
        print("Some sound")

class Dog(Animal):
    def __init__(self, name: str):
        super().__init__(name)

dog = Dog("Test")
dog.speak()
```

2. Method overriding

Modify `Dog` so that:

* `speak()` prints `"Woof!"` instead of `"Some sound"`

```python
class Dog(Animal):
    def __init__(self, name: str):
        super().__init__(name)

    def speak(self):
        print("Woof!")

dog = Dog("Test")
dog.speak()
```

6. Multiple methods inheritance

Create:

* `Shape` with method `describe()` → `"I am a shape"`

Create `Rectangle`:

* attributes: width, height
* method `area()`

```python
class Shape():
    def descirbe(self):
        print("This is a shape")

class Rectangle(Shape):
    def __init__(self, width: int, height: int):
        self.width = width
        self.height = height

    def descirbe(self):
        print("This is a rectangle")

    def area(self):
        return self.width * self.height

rec = Rectangle(15, 12)
rec.descirbe()
rec.area()
```

7. Override + new behavior

Create:

* `Employee(name, salary)`
* method `work()` → `"Working..."`

Create `Manager`:

* override `work()` → `"Managing team"`

```python
class Employee:
    def __init__(self, name: str, salary: int):
        self.name = name
        self.salary = salary

    def work(self):
        print("Working ... ")

class Manager(Employee):
    def work(self):
        print("Managing ... ")

e = Employee("John", 100)
e.work()

m = Manager("Joey", 200)
m.work()
```

8. Add functionality to parent method

Create:

* `BankAccount(balance)`
* method `deposit(amount)`

Create `SavingsAccount`:

* inherits from `BankAccount`
* adds method `add_interest(rate)`
* increases balance using interest

```python
class BankAccount:
    def __init__(self, amount: int):
        self.amount = amount

    def deposit(self, amount: int):
        self.amount += amount

class SavingsAccount(BankAccount):
    def add_intrest(self, rate: float):
        self.amount += rate * self.amount

s = SavingsAccount(1000)
print(s.amount)
s.deposit(100)
print(s.amount)
s.add_intrest(0.01)
print(s.amount)
```
