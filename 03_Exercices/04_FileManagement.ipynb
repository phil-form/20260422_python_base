4. Parse the data into a dictionary

````python
data = []

with open('data/people.csv', 'r') as f:
    for line in f:
        line = line.strip()
        if line:
            name, age = line.split(",")

            data.append({
                "name": name,
                "age": int(age)
            })

print(data)
````

5. Retrive the average age and the oldest person

````python
oldest = data[0]

for person in data:
    if person['age'] > oldest['age']:
        oldest = person

print(oldest)
````

6. Retrieve only the adults in the file and put them into a new file named adult.csv

````python
adults = []

for person in data:
    if person['age'] > 18:
        adults.append(person)

print(adults)

with open('data/adults.csv', 'w') as f:
    for person in adults:
        f.write(f'{person["name"]},{person["age"]}\n')

````

8. Write the content of adult.csv into a json file

````python
import json

adults = []

with open('data/adults.csv', 'r') as f:
    for line in f:
        line = line.strip()
        if line:
            name, age_str = line.split(",")
            age = int(age_str)

            if age > 18:
                adults.append({
                    "name": name,
                    "age": age
                })

with open('data/adult.json', 'w') as f:
    json.dump(adults, f)
````

9. Read the person.csv and retrieve the following informations :

* Total people
* Average age
* Youngest and oldest

And write those data into report.txt

````python
data = []

with open('data/people.csv', 'r') as f:
    for line in f:
        line = line.strip()

        if line:
            name, age_str = line.split(",")
            data.append({
                "name": name,
                "age": int(age_str)
            })

oldest = data[0]
youngest = data[0]
total_age = 0
total_person = len(data)
for person in data:
    if person['age'] < youngest['age']:
        youngest = person

    if person['age'] > oldest['age']:
        oldest = person

    total_age += person['age']

average_age = total_age / total_person

with open('data/summary.txt', 'w') as f:
    f.write(f'Average age: {average_age}\n')
    f.write(f'Oldest : {oldest['name']}\n')
    f.write(f'Youngest : {youngest['name']}\n')
    f.write(f'Number of data: {total_person}\n')
````
