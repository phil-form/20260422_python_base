# 4) If/Elif/Else

* Create a small algorithm that takes a year as an input and tells if the year is a leap year or not. To recon if it is a leap year or not leap years are every 4 years, but not every 100 years except every 400 years.

````python
year = int(input("Enter a year: "))

if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
    print(year , " is a leap year.")
else:
    print(year , " is not a leap year.")
````

# Collection Medium

## 1. Count even numbers

Given:

```python
numbers = [1, 2, 3, 4, 5, 6]
```

Count how many numbers are even and print the result.

````python
numbers = [1, 2, 3, 4, 5, 6]

even_counter = 0
for elem in numbers:
    if elem % 2 == 0:
        even_counter += 1

print(even_counter)
````

## 2. Sum of positive numbers

```python
numbers = [-2, 5, -1, 7, -3]
```

Calculate the sum of only the **positive numbers**.

````python
numbers = [-2, 5, -1, 7, -3]

total = 0
for elem in numbers:
    if elem > 0:
        total += elem

print(total)
````

## 3. Find the largest number

```python
numbers = [3, 8, 2, 10, 5]
```

Find and print the **largest number** without using `max()`.

````python
numbers = [3, 8, 2, 10, 5]

# max(numbers)

max_nbr = numbers[0]

for elem in numbers:
    if elem > max_nbr:
        max_nbr = elem

for elem in numbers[1:]:
    if elem > max_nbr:
        max_nbr = elem

print(max_nbr)
````

## 5. Count occurrences

```python
letters = ["a", "b", "a", "c", "a"]
```

Count how many times `"a"` appears.

````python
letters = ["a", "b", "a", "c", "a"]

count_occur = {}

for elem in letters:
    if elem in count_occur.keys():
        count_occur[elem] += 1
    else:
        count_occur[elem] = 1

print(count_occur)
````

Create a new collection with **no duplicates** and print it.

````python
numbers = [1, 2, 2, 3, 4, 4]

print(set(numbers))

arr = []
for elem in numbers:
    if not elem in arr:
        arr.append(elem)

print(arr)
````

## 8. Count vowels in a string

```python
text = "hello world"
```

Count how many vowels (`a, e, i, o, u`) are in the string.

````python
text = "hello world"
vowels = ["a", "e", "i", "o", "u"]

vowels_count = 0
for c in text:
    if c in vowels:
        vowels_count += 1

print(vowels_count)
````

## 9. Find common elements

```python
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]
```

Print the numbers that appear in **both lists**.

````python
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]

common_elem = []
for elem in list1:
    for elem2 in list2:
        if elem == elem2:
            print(elem)
            common_elem.append(elem)

print(common_elem)

common_elem = []
for elem in list1:
    if elem in list2:
        common_elem.append(elem)

print(common_elem)
````
