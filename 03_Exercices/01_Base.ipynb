## 2) Operators

* Do an algorithm that convert a number of seconds entered by the user into "XX DAYS XX HOURS XX MIN XX SEC"

```python
total = int(input("Enter a number of seconds: "))

days = total // (60 * 60 * 24)
remaining = total % (60 * 60 * 24)
hours = remaining // (60 * 60)
remaining = remaining % (60 * 60)
minutes = remaining // (60)
seconds = remaining % (60)
print(total, "s is : ", days, " Days, ", hours, " Hours, ", minutes, " Min, ", seconds, " Seconds.")
```

# 4) If/Elif/Else

* Create a small algorithm that takes a year as an input and tells if the year is a leap year or not. To recon if it is a leap year or not leap years are every 4 years, but not every 100 years except every 400 years.

```python
year = int(input("Enter a year: "))

if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
    print(year , " is a leap year.")
else:
    print(year , " is not a leap year.")
```
