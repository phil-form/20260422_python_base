## 3. Rectangle class

Create a class `Rectangle` with:

* attributes: `width`, `height`
* method: `area()` → returns area
* method: `perimeter()` → returns perimeter

````python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return (self.width + self.height) * 2

r = Rectangle(15, 20)
print(r.area())
print(r.perimeter())
````

## 4. Student class

Create a class `Student` with:

* attributes: `name`, `grades` (list)
* method: `average()` → returns average grade

````python
class Student:
    def __init__(self, name: str, grades: list):
        self.name = name
        self.grades = grades

    def average(self):
        return sum(self.grades) / len(self.grades)

s = Student("Test", [10, 15, 3, 15, 18, 17])

print(s.average())
````

Create a class `BankAccount` with:

* attribute: `balance`

  * `deposit(amount)`

````python
class BankAcount:
    def __init__(self, balance: float):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

    def show_balance(self):
        print(self.balance)

ba = BankAcount(100)

ba.show_balance()
ba.deposit(50)
ba.show_balance()
ba.withdraw(20)
ba.show_balance()
````

## 7. Counter class

Create a class `Counter` with:

* attribute: `value` (starts at 0)
* methods:

  * `increment()`
  * `decrement()`
  * `reset()`

````python
class Counter:
    def __init__(self, start_val = 0):
        self.val = start_val

    def increment(self):
        self.val += 1

    def decrement(self):
        self.val -= 1

    def reset(self):
        self.val = 0

c = Counter()

print(c.val)
c.increment()
print(c.val)
c.increment()
c.increment()
print(c.val)
c.decrement()
print(c.val)
c.increment()
c.increment()
c.increment()
print(c.val)
c.reset()
print(c.val)
````

## 9. Temperature class

Create a class `Temperature` with:

* attribute: `celsius`
* methods:

  * `to_fahrenheit()`
  * `is_hot()` → returns True if > 30°C

````python
class Temperature:
    def __init__(self, temp):
        self.temp = temp

    def to_fahrenheit(self):
        return (self.temp * 9 / 5) + 32

    def is_hot(self):
        return self.temp > 30

temp = Temperature(30)

print(temp.to_fahrenheit())
print(temp.is_hot())
````

Extend the same `Book` class and implement:

```python

```

so it returns:


## 4. Equality check (`__eq__`)

Create a class `Student`:

* attributes: `name`, `grade`

Implement:

```python
__eq__
```

so two students are equal if their grades are equal.

````python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def __eq__(self, other):
        if isinstance(other, Student):
            return self.grade == other.grade

        return False

st1 = Student("Test", 10)
st2 = Student("Asdf", 20)
print(st1 == st2)
st3 = Student("Qwer", 10)
print(st1 == st3)
````

Extend `Student`:
Implement:

```python
__lt__
```

so students can be compared by grade.

👉 Example:

```python
s1 < s2
```

````python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def __eq__(self, other):
        if isinstance(other, Student):
            return self.grade == other.grade

    def __lt__(self, other):
        if isinstance(other, Student):
            return self.grade < other.grade

        return False

st1 = Student("Test", 10)
st2 = Student("Asdf", 20)
print(st1 < st2)
st3 = Student("Qwer", 10)
print(st1 < st3)
````

## 7. Indexing support (`__getitem__`)

Create a class `Library`:

* attribute: `books` (list)

Implement:

```python
__getitem__
```

so you can do:

```python
library[0]
```

````python
class Library:
    def __init__(self, books):
        self.books = books

    def __getitem__(self, item):
        return self.books[item]

    def __len__(self):
        return len(self.books)

lib = Library(["LOTR", "The Expense", "Master And Commander"])

lib[0]
````

## 8. Assignment by index (`__setitem__`)

Extend `Library`:
Implement:

```python
__setitem__
```

so you can replace a book:

```python
library[0] = new_book
```

````python
class Library:
    def __init__(self, books):
        self.books = books

    def __getitem__(self, item):
        return self.books[item]

    def __setitem__(self, key, value):
        self.books[key] = value

    def __len__(self):
        return len(self.books)

lib = Library(["LOTR", "The Expense", "Master And Commander"])

lib[0] = "The Far side of the world"
print(lib[0])
````

## 9. Custom iterator (`__iter__`)

Create a class `Countdown`:

* attribute: `start`

Implement:

```python
__iter__
```

so you can loop:

```python
for i in Countdown(5):
    print(i)
```

Output:

```
5 4 3 2 1
```

````python
class Countdown:
    def __init__(self, n):
        self.n = n
        self.__count = n

    def __iter__(self):
        self.__count = self.n
        return self

    def __next__(self):
        if(self.__count == 0):
            raise StopIteration

        val = self.__count
        self.__count -= 1

        return val

for i in Countdown(5):
    print(i)
````
