## 1. Simple Person class

## 2. Dog class

Create a class `Dog` with:

* attributes: `name`, `breed`
* method: `bark()` → prints `"Woof!" + name`

```python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):
        print(self.name + " woof!")

d = Dog("Test", "German-shepherd")
print(d.name)
print(d.breed)
d.bark()
```

## 3. Rectangle class

Create a class `Rectangle` with:

* attributes: `width`, `height`
* method: `area()` → returns area
* method: `perimeter()` → returns perimeter

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return (self.width + self.height) * 2

r = Rectangle(15, 20)
print(r.area())
print(r.perimeter())
```

## 4. Student class

Create a class `Student` with:

* attributes: `name`, `grades` (list)
* method: `average()` → returns average grade

```python
class Student:
    def __init__(self, name: str, grades: list):
        self.name = name
        self.grades = grades

    def average(self):
        return sum(self.grades) / len(self.grades)

s = Student("Test", [10, 15, 3, 15, 18, 17])

print(s.average())
```

## 5. Bank account

Create a class `BankAccount` with:

* attribute: `balance`
* methods:

  * `deposit(amount)`
  * `withdraw(amount)`
  * `show_balance()`

```python
class BankAcount:
    def __init__(self, balance: float):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

    def show_balance(self):
        print(self.balance)

ba = BankAcount(100)

ba.show_balance()
ba.deposit(50)
ba.show_balance()
ba.withdraw(20)
ba.show_balance()
```

## 6. Car class

Create a class `Car` with:

* attributes: `brand`, `speed`
* methods:

  * `accelerate(amount)`
  * `brake(amount)`

```python
class Car:
    def __init__(self, brand, speed = 0):
        self.brand = brand
        self.speed = speed

    def accelerate(self, amount):
        self.speed += amount

    def brake(self):
        self.speed = 0

c = Car("C1", 0)

print(c.speed)
c.accelerate(15)
print(c.speed)
c.accelerate(20)
print(c.speed)
c.brake()
print(c.speed)
```

## 7. Counter class

Create a class `Counter` with:

* attribute: `value` (starts at 0)
* methods:

  * `increment()`
  * `decrement()`
  * `reset()`

```python
class Counter:
    def __init__(self, start_val = 0):
        self.val = start_val

    def increment(self):
        self.val += 1

    def decrement(self):
        self.val -= 1

    def reset(self):
        self.val = 0

c = Counter()

print(c.val)
c.increment()
print(c.val)
c.increment()
c.increment()
print(c.val)
c.decrement()
print(c.val)
c.increment()
c.increment()
c.increment()
print(c.val)
c.reset()
print(c.val)
```

## 9. Temperature class

Create a class `Temperature` with:

* attribute: `celsius`
* methods:

  * `to_fahrenheit()`
  * `is_hot()` → returns True if > 30°C

```python
class Temperature:
    def __init__(self, temp):
        self.temp = temp

    def to_fahrenheit(self):
        return (self.temp * 9 / 5) + 32

    def is_hot(self):
        return self.temp > 30

temp = Temperature(30)

print(temp.to_fahrenheit())
print(temp.is_hot())
```

## 10. Simple shopping item

Create a class `Item` with:

* attributes: `name`, `price`, `quantity`
* method: `total_price()` → returns price × quantity

```python
class Item:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity

i = Item("Orange", 2, 10)
print(i.total_price())
```
