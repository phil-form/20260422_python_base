```python
def add(a, b):
    return a + b

# If you don't know or want to allow any number of arguments in the function you can use
# the "*" before the argument name.
# The all arguments will be passed in the "args" variable as an list
# e.g. the print(...) function, I don't know how many arguments the user might want to pass
#      so it uses the *args to say that the user might use as many arguments as he wants.
def add_multiple(*args):
    addition = 0
    for val in args:
        addition += val

    return addition

print(add_multiple(5, 5, 7, 25, 36, 55))

# The "**" works similarly to the "*" but here instead of a list of values that is passed to the function,
# it is a dictionary.
# This can be useful when connecting to a database :
# table_name.find_by(column_name=value, column_name2=value2)
# that would create a query like SELECT * FROM table_name WHERE column_name = 'value' AND column_name2 = 'value2'
def example_kwarg(**kwargs):
    print(kwargs)

example_kwarg(test=5, asdf=9)
```

```python
def typing(arg1: int, arg2: float, arg3: str) -> str:
    print(arg1, arg2, arg3)

typing("asdasdasda", 12.0, "qwe")

typing(arg3=3, arg1=2, arg2="asd")
```
