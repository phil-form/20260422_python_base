```python
def int_input(msg):
    val = None
    while val is None:
        try:
            val = int(input(msg))
        except Exception as e:
            print("Enter a valid number")

    return val


def int_input2(msg):
    try:
        return int(input(msg))
    except Exception as e:
        print("Enter a valid number")
        return int_input(msg)

def int_input3(msg):
    while True:
        try:
            return int(input(msg))
        except Exception as e:
            print("Enter a valid number")

val = int_input("Enter a number: ")
print(val)
```
