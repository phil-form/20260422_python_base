# List

```python
# Create empty list
my_list = []
my_list = list()

# Create with values
my_list = [1, 2, 3]
my_list = list((1, 2, 3))

# Access values
print(my_list[0])
print(my_list[1])
print(my_list[2])

# Add a value
my_list.append(4)
print(my_list)

# remove a value
my_list.remove(2)
print(my_list)

# Remove index
my_list.pop(2)
print(my_list)

# Know the length of a list
length = len(my_list)
print(length)

# Clear
my_list.clear()
print(my_list)
```

# Dictionary

```python
# Create empty dictionary
my_dict = {}
my_dict = dict()

print(my_dict)

# Create with values :
my_dict = {
    "firstname": "John",
    "lastname": "Doe",
}
my_dict = dict(firstname="John", lastname="Doe")
print(my_dict)

# Access items
print(my_dict["firstname"])
print(my_dict["lastname"])

# Modify a value at key
my_dict["firstname"] = "Jane"
print(my_dict)

# Delete key
my_dict.pop("firstname")
print(my_dict)

# Clear
my_dict.clear()
print(my_dict)
```
