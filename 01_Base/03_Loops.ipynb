# While loops

For all loops pay attention to have an exit condition (so that the condition in the while, is not met at some point).

```python
nbr = 0

while nbr < 1 or nbr > 10:
    nbr = int(input("Enter a number from 1 to 10 : "))

print(nbr)
```

# For loop

```python
for i in range(10):
    print(i)
```

```python
for i in range(10, 15):
    print(i)
```

```python
for i in range(10, 15, 2):
    print(i)
```
